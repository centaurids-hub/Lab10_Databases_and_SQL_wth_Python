```python
# PROCEDURE 1: Create and Connect to a SQLite Database
import sqlite3

conn = sqlite3.connect("students.db")
cursor = conn.cursor()

print("Connected to students.db")

# Try It
print(type(conn), type(cursor))
```

```python
# PROCEDURE 2: Create a Table
create_table_query = """
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    full_name TEXT NOT NULL,
    course TEXT NOT NULL,
    year_level INTEGER NOT NULL
)
"""

cursor.execute(create_table_query)
conn.commit()
print("Table 'students' created successfully.")

# Try It
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
print("Tables in database:", cursor.fetchall())
```

```python
# PROCEDURE 3: Insert Sample Data into the Table
students = [
    ("Jhonren Traballo", "BS ECE", 3),
    ("Ruselle Laude", "BS EE", 2),
    ("Rodrigo Balayo", "BS CE", 1)
]

insert_query = """
INSERT INTO students (full_name, course, year_level)
VALUES (?, ?, ?)
"""

cursor.executemany(insert_query, students)
conn.commit()
print("Sample data inserted into 'students' table.")

# Try It
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# PROCEDURE 4: Query the Data

# All students
cursor.execute("SELECT * FROM students")
print("All Students:")
for row in cursor.fetchall():
    print(row)

# Only BS ECE
cursor.execute("SELECT * FROM students WHERE course = 'BS ECE'")
print("\nBS ECE Students:")
print(cursor.fetchall())

# Sorted by year level descending
cursor.execute("SELECT * FROM students ORDER BY year_level DESC")
print("\nSorted by Year Level:")
print(cursor.fetchall())

# Top 2 only
cursor.execute("SELECT * FROM students LIMIT 2")
print("\nTop 2 Students:")
print(cursor.fetchall())

# Try It
cursor.execute("SELECT * FROM students WHERE course = 'BS CE' AND year_level >= 1")
print("\nFiltered list of BS CE students:")
print(cursor.fetchall())
```

```python
# PROCEDURE 5:  Update and Delete Record

# Update Isaac Newton to 4th year
cursor.execute("""
UPDATE students
SET year_level = 4
WHERE full_name = 'Isaac Newton'
""")
conn.commit()
print("Updated Isaac Newton to 4th year.")

# Delete Rjay Canoza
cursor.execute("""
DELETE FROM students
WHERE full_name = 'Rjay Canoza'
""")
conn.commit()
print("Deleted Rjay Canoza from the database.")

# Try It - Update Mika Santos to 2nd year
cursor.execute("""
UPDATE students
SET year_level = 2
WHERE full_name = 'Sean Kaylee Manzano'
""")
conn.commit()
print("Sean Kaylee Manzano has been updated to 2nd year.")
```

```python
# PROCEDURE 6: Use Aggregate Functions

cursor.execute("SELECT COUNT(*) FROM students")
print("Total Students:", cursor.fetchone()[0])

cursor.execute("SELECT AVG(year_level) FROM students")
print("Average Year Level:", cursor.fetchone()[0])

cursor.execute("SELECT course, COUNT(*) FROM students GROUP BY course")
print("Students per Course:", cursor.fetchall())

# Try It
cursor.execute("SELECT course, COUNT(*) FROM students GROUP BY course")
print("\nTotal number of students per course:")
print(cursor.fetchall())
```

```python
# PROCEDURE 7: Close the Database Connection
cursor.close()
conn.close()
print("Database connection closed.")
```

```python
# PROCEDURE 8: Insert Data via User Input
# Re-open connection first if you ran Procedure 8
conn = sqlite3.connect("students.db")
cursor = conn.cursor()

full_name = input("Enter full name: ")
course = input("Enter course (e.g., BS ECE): ")
year_level = int(input("Enter year level (e.g., 1, 2, 3): "))

cursor.execute("""
INSERT INTO students (full_name, course, year_level)
VALUES (?, ?, ?)
""", (full_name, course, year_level))
conn.commit()
print("Student record successfully added.")

# Verify it was added
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(row)
```

```python
# PROCEDURE 10: Export Query Results to CSV File
import csv

cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

with open("students_export.csv", "w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["ID", "Full Name", "Course", "Year Level"])
    writer.writerows(rows)

print("Data exported to 'students_export.csv'.")
```

```python
# PROCEDURE 11: Create and Join Tables

# Create the courses table
cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course_code TEXT PRIMARY KEY,
    description TEXT NOT NULL
)
""")
conn.commit()

# Insert course data
course_data = [
    ("BS ECE", "Electronics and Communications Engineering"),
    ("BS EE", "Electrical Engineering"),
    ("BS COE", "Computer Engineering"),
    ("BS CE", "Civil Engineering")
]

cursor.executemany("""
INSERT OR IGNORE INTO courses (course_code, description)
VALUES (?, ?)
""", course_data)
conn.commit()

# JOIN query
cursor.execute("""
SELECT s.full_name, s.course, c.description
FROM students s
JOIN courses c ON s.course = c.course_code
""")

for row in cursor.fetchall():
    print(row)
```

```python
# PROCEDURE 12: Filtered and Sorted JOIN Query

# Filter by BS CE, sorted alphabetically
cursor.execute("""
SELECT s.full_name, s.course, c.description
FROM students s
JOIN courses c ON s.course = c.course_code
WHERE s.course = 'BS CE'
ORDER BY s.full_name ASC
""")

for row in cursor.fetchall():
    print(row)

# Try It - Filter by BS ECE instead
cursor.execute("""
SELECT s.full_name, s.course, c.description
FROM students s
JOIN courses c ON s.course = c.course_code
WHERE s.course = 'BS ECE'
ORDER BY s.full_name ASC
""")

print("\nBS ECE students sorted by name:")
for row in cursor.fetchall():
    print(row)
```

```python
# PROCEDURE 13: Search Using Pattern Matching (LIKE)

# Part A - Static search
cursor.execute("""
SELECT * FROM students
WHERE full_name LIKE '%Reyes%'
""")
print("Static search result:")
for row in cursor.fetchall():
    print(row)

# Part B - Dynamic search with user input
keyword = input("Enter a name keyword to search: ")
cursor.execute("""
SELECT * FROM students
WHERE full_name LIKE ?
""", ('%' + keyword + '%',))

print("Search results:")
for row in cursor.fetchall():
    print(row)
```
